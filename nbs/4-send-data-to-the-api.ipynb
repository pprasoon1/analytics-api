```python
path = "/api/events/"
base_url = "http://localhost:8000"
create_endpoint = f"{base_url}{path}"

response = requests.post(create_endpoint, json={"page":"/test+", }) #HTTP POST
print(response.headers)
print('ok', response.ok)
if response.ok:
    data = response.json()
    print(data, type(data), data.get('items'))
else:
    print(response.text) 
```
